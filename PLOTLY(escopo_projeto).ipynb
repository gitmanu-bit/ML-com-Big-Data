```python
import pandas as pd
import plotly.graph_objects as go

# Carrega os dados do arquivo CSV
#'latin-1' Serve para fazer o colab fazer a leitura dos caracteres do documento
try:
    df = pd.read_csv('/content/media_2022_2025(Sheet1).csv', encoding='latin-1')
except UnicodeDecodeError:
    df = pd.read_csv('/content/media_2022_2025(Sheet1).csv', encoding='latin-1')

# Cria um gráfico de linha usando graph_objects
fig = go.Figure()

# Adiciona uma trace para cada variável y
# The x-axis should use the column containing the time series data, likely 'Unnamed: 0'
fig.add_trace(go.Scatter(x=df["Data"], y=df["Óxido de Enxofre - SO2 (µg/m³)"], mode='lines', name='Óxido de Enxofre - SO2 (µg/m³)'))
fig.add_trace(go.Scatter(x=df["Data"], y=df["Monóxido de carbono  - CO (ppm)"], mode='lines', name='Monóxido de carbono  - CO (ppm)'))
fig.add_trace(go.Scatter(x=df["Data"], y=df["Ozônio -O3 (µg/m³)"], mode='lines', name='Ozônio -O3 (µg/m³)'))

# Atualiza o layout do gráfico (opcional)
fig.update_layout(
    title='Medição SEUMA 2022-2025',
    xaxis_title='Anos e Meses',  # Pode ser renomeado se "Unnamed: 0" for uma coluna de tempo ou data
    yaxis_title='Concentração',
    hovermode='x unified' # Melhora a visualização do hover tooltip
)

# Mostra o gráfico
fig.show()
# fig = px.scatter(df, x="x", y="y", color="blue")
# fig.show()
```

```python
# Carregar o arquivo CSV
# 'latin-1' Serve para fazer o colab fazer a leitura dos caracteres do documento
df = pd.read_csv("/content/media_2022_2025(Sheet1).csv", encoding='latin-1')

# Converter a coluna de data explicitamente informando o formato correto
# O formato parece ser M/D/AAAA com base na mensagem de erro, então use '%m/%d/%Y'
df['Data'] = pd.to_datetime(df['Data'], format='%m/%d/%Y')

# Criar colunas de ano e mês com nomes em português
df['Ano'] = df['Data'].dt.year
meses_pt = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
            'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

# "Lambda" Serve para fazer uma declaração que vai ter um valor em específico
# "x - 1" Serve para mostrar ao colab que tem uma ordem de meses a seguir, começando do mês 1
df['Mês'] = df['Data'].dt.month.apply(lambda x: meses_pt[x - 1])

# "ordered=True" Serve para mostrar ao colab que ele tem uma ordem a seguir, já declarada acima
df['Mês'] = pd.Categorical(df['Mês'], categories=meses_pt, ordered=True)

# Agrupar por ano e mês e calcular a média
df_grouped = df.groupby(['Ano', 'Mês']).mean(numeric_only=True).reset_index()

# Criar gráfico
grafico = go.Figure()
poluentes = ['Óxido de Enxofre - SO2 (µg/m³)', 'Ozônio -O3 (µg/m³)', 'Monóxido de carbono  - CO (ppm)']

for poluente in poluentes:
    for ano in sorted(df_grouped['Ano'].unique()):
        dados_ano = df_grouped[df_grouped['Ano'] == ano]
        grafico.add_trace(go.Bar(
            x=dados_ano['Mês'],
            y=dados_ano[poluente],
            name=f"{poluente} - {ano}"
        ))

grafico.update_layout(
    title='Média Mensal dos Poluentes (2022-2025)',
    xaxis_title='Meses',
    yaxis_title='Concentração Média',
    barmode='group'
)

grafico.show()
```

```python
import pandas as pd

# Carregar o CSV
#"encoding='latin-1" Serve para fazer o colab fazer a leitura dos caracteres do documento
df = pd.read_csv("/content/media_2022_2025(Sheet1).csv", encoding='latin-1')
#Alterar data para o tipo certo
df['Data'] = pd.to_datetime(df['Data'], format='%m/%d/%Y')

# Poluentes
poluentes = ['Óxido de Enxofre - SO2 (µg/m³)','Ozônio -O3 (µg/m³)', 'Monóxido de carbono  - CO (ppm)']

# Médias anuais
df['Ano'] = df['Data'].dt.year
medias_anuais = df.groupby('Ano')[poluentes].mean()

# Mostrar os resultados
print("📊 Crescimento percentual ano a ano:")

# Crescimento percentual ano a ano
crescimento_percentual = medias_anuais.pct_change() * 100
# Crescimento médio anual
media_crescimento_anual = crescimento_percentual.mean()

# Projeção para 5 anos (crescimento composto)
projecao_5_anos = ((1 + media_crescimento_anual / 100) ** 5 - 1) * 100

# Exibir resultados
#round(2) Serve para retirar as casas décimais (no caso deixar apenas duas casas apoś o ponto)
print("Crescimento percentual ano a ano:")
print(crescimento_percentual.round(2))
print("\nCrescimento médio anual:")
print(media_crescimento_anual.round(2))
print("\nProjeção de crescimento para 5 anos (se a tendência continuar):")
print(projecao_5_anos.round(2))
```
